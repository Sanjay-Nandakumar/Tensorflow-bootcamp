```
#Matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
x = np.array([1,2,3])
y = x*2
plt.plot(x,y)
plt.show()
```

```
plt.title('title')
plt.ylabel("Y axis")
plt.xlabel("X axis")
plt.show()
```

```
plt.scatter(x,y)
```

```
plt.ylim(0,100)
```

```
plt.plot(x,y)
```

```
plt.ylim(0,100)
plt.plot(x,y)
```

```
plt.plot(x,y,marker='o')
```

```
plt.plot(x,y,marker='o',markersize=15)
```
